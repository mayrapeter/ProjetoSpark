```python
import csv
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import pandas as  pd
```

# CLUSTER

Constrindo a WordCloud para Apple sozinha

```python
df_apple = pd.read_csv('apple.csv')
apple = {}
```

```python
for i,row in df_apple.iterrows():
    apple[df_apple['Palavra'][i]] = df_apple['Relev√¢ncia'][i]
```

```python
wordcloud = WordCloud(background_color="white")
wordcloud.generate_from_frequencies(apple)
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```
